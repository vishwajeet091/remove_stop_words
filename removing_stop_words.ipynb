```python
#NLTK method
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

def remove_stop_words_nltk(text):
  words=word_tokenize(text)
  stopword=set(stopwords.words('english'))
  filtered_words=[word for word in words if word.lower() not in stopword]
  filtered_text=' '.join(filtered_words)
  return filtered_text

df['Summary']=df['Summary'].apply(lambda x:str(x))
df['nltk_stoppped_words']=df['Summary'].apply(lambda x:remove_stop_words_nltk(x))
df
```

```python
#SpaCy
import spacy
nlp=spacy.load("en_core_web_sm")

def remove_stop_words_spacy(text):
  doc=nlp(text)
  filtered_words=[token.text for token in doc if not token.is_stop]
  filtered_text=' '.join(filtered_words)
  return filtered_text

df['Summary']=df['Summary'].apply(lambda x:str(x))
df['spacy_stoppped_words']=df['Summary'].apply(lambda x:remove_stop_words_nltk(x))
df
```

```python
#Scikit-learn
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

def remove_stop_words_sklearn(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in ENGLISH_STOP_WORDS]
    filtered_text = ' '.join(filtered_words)
    return filtered_text


df['Summary']=df['Summary'].apply(lambda x:str(x))
df['remove_stop_words_sklearn']=df['Summary'].apply(lambda x:remove_stop_words_nltk(x))
df
```

```python
def remove_stop_words_custom(text, custom_stop_words):
    words = text.split()
    custom_stop_word=[word.lower() for word in custom_stop_words]
    filtered_words = [word for word in words if word.lower() not in custom_stop_word]
    filtered_text = ' '.join(filtered_words)
    return filtered_text

df['Summary']=df['Summary'].apply(lambda x:str(x))
df['remove_stop_words_custom']=df['Summary'].apply(lambda x:remove_stop_words_nltk(x))
df
```
